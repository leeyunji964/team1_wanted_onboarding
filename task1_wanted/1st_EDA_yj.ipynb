# 연매출액 기준 EDA

- 연매출액 증가율 ex) 2019 증가율 : (2019-2018)/2018*100(%)
    - 해당 EDA를 하기 위해서는 최소 2년 이상의 데이터가 있어야 함
    - 특정 기업 고른 후 진행

```python
def sales(x,y):
    plt.rcParams['figure.figsize'] = [15,5]

    plt.plot(x,y)
    plt.xticks(rotation=-45)
    #plt.legend(loc='upper left')
    plt.show()
```

```python
a = data.groupby('회사ID').count()
# 월이 24개 이상인 것 찾기
a
```

```python
a.reset_index(inplace=True)
over_years = a.loc[a['월'] >= 24] #최소 24개월(2년) 이상의 데이터만 추출
over_years
```

```python
#over_years.reset_index(inplace=True)
over_years = over_years.drop('index', axis=1)
over_years
```

```python
id_list = over_years['회사ID'].unique().tolist()
# data에서 해당 id만 추출
growth = data.loc[data['회사ID'].isin(id_list)]

growth.reset_index(inplace=True)
growth = growth.drop('index', axis=1)
growth # 최소 2년 이상
```

```python
growth.info() #년도,월 = int => string
```

```python
growth['년도'] = growth['년도'].astype(str)
growth['월'] = growth['월'].astype(str)
```

```python
print(growth['회사ID'].nunique()) # 최소 24개월 이상 데이터 보유 회사는 2193개
print(growth['회사ID'].unique()) # 우선 14271, 44851, 59533,514934, 515264, 516653 확인
```

```python
growth
```

```python
growth['회사ID'].unique().tolist()[:10]
```

```python
cor_id = growth['회사ID'].unique().tolist()[:10]

for i in range(len(cor_id)):
    d = growth.loc[growth['회사ID'] == cor_id[i]]
    plt.title("회사ID " + str(cor_id[i]) + "의 연매출액", fontsize=15)

    sales(d['년도'],d['연매출액'])    
```
